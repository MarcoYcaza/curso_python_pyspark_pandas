### Condicionales

```python
from pyspark.sql.functions import when
from pyspark.sql.functions import lit
```

```python
cnd1        = (col("transaction_amount") > 70000) & (col("transaction_amount") < 80000)  
cnd2        = (col("transaction_amount") > 10000) & (col("transaction_amount") < 30000) 

columnExpr = when(cnd1,lit("MuyAlto")).\
             when(cnd2,lit("Alto")).\
             otherwise(lit("")) 

df.withColumn("Calificativo" ,columnExpr).show(2)
```
