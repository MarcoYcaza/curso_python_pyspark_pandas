## Compresión de listas

```python
lista = []

cadena = "AAABCCDAAAABCCAAAAUUUIIAAAA"
for letra in cadena:
  if letra == "A":
    lista.append(letra)

print(lista)
```

```python
import string

cadena = "$M%AR&C#""$#......ASDB/%$SM"

punctuaciones = string.punctuation

#respuesta = [ f(x) for x in <>  if condicion]
#print(respuesta)

# "A" not in ["B","C","A"] -> False
# "A" not in ["B","C"]     -> True
# "A" not in "BC"          -> True

respuesta = [ caracter.lower() for caracter in cadena if caracter not in punctuaciones]

nueva_cadena = "".join(respuesta)

print(nueva_cadena)
```

```python
import string

cadena = "$M%AR&C#""$#......ASDB/%$SM"

punctuaciones = string.punctuation

# "A" not in "BC"          -> True

is_punctuation = lambda _ : _ not in punctuaciones

list(filter(is_punctuation,cadena))
```
