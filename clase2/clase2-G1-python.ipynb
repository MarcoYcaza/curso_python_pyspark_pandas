## Ejemplo de herencia

````python
class animal:

  def __init__(self , tipo , genoma):

    self.tipo = tipo
    self.genoma = genoma

  def decirGenoma(self):
    print(f"Mi genoma es el siguiente {self.genoma}")

class cat(animal):

  def __init__(self, tipo , genoma ,nombre,raza):

    self.tipo = tipo
    self.genoma = genoma

    self.nombre = nombre
    self.raza = raza

    super().__init__(self.tipo , self.genoma)


  def maullar(self):
    print("Estoy maullando!")

class loro(animal):

  def __init__(self, tipo , genoma ,nombre,raza):

    self.tipo = tipo
    self.genoma = genoma

    self.nombre = nombre
    self.raza = raza

    super().__init__(self.tipo , self.genoma)

  def volar(self):

    print(f" Mi nombres es {self.nombre} y estoy volando")


cat1 = cat("mamífero"," AGTCUAGTCUUTGA" ,"felix" , "siames")


loro1 = loro("ave" , "TUGTUGUAAACC" , "pepe" , "Cacatúa")
````

````python
cat1.decirGenoma()
cat1.maullar()
````

````python
loro1.decirGenoma()
loro1.volar()
````
