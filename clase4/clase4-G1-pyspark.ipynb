### RDD a partir de archivo "inversiones.csv"

```python
miRDD = spark.sparkContext.textFile("inversiones.csv")
```

```python
type(miRDD)
```

```python

map_fn = lambda _ : (_.split(",")[3] , 1)

redux_fn = lambda a,b : a + b 


miRDD.map(map_fn).collect()

miRDD2 = miRDD.map(map_fn).reduceByKey(redux_fn)
```

```python
lista = [23,24,25,67,89,43]

# suma embebida
sum(lista)

#suma más 'computacional'

import functools
functools.reduce(lambda a, b: a+b, lista)
```
