```python
recurso_iterable = [0,1,2,3,4,5,6,7]

#             [  f(elemento)   por elemento   en   iterable  si condicion] 

nueva_lista = [ i*2 for i in  recurso_iterable]

print(nueva_lista)
```

```python
                                              # [0,1,2,3,4,5,6,7]  
nueva_lista = [ x*2 for x in  recurso_iterable if (x > 3) ]

print(nueva_lista)
```

```python
            #      <-------f(x)-------->      x         iterable            
            #      <------------------->      x         iterable  
nueva_lista = [    (2*x if x%2==0 else 47)  for x in  recurso_iterable ]

print(nueva_lista)
```

```python
from functools import reduce
```

```python
my_function = lambda a,b : a+b

lista = [1,2,3,4,5,8,72,2,6,8]

resultado = reduce(my_function , lista)

print(resultado)
```

```python
lista = [1,2,3,4,5,8,72,2,6,8]

resultado = reduce( lambda a,b : a+b , lista)

print(resultado)

#------------------

print(sum(lista))
```

```python
class Dog:

  def __init__(self ,name,edad):
     self.Nombre = name
     self.Edad = edad
  
  def ladrar(self):
    print(f"Woof! Woof  Soy {self.Nombre} ")
  
  def saludar(self):
    print(f"Tengo {self.Edad }  años de edad")
```

```python
dog1 = Dog("Sizzy", 2)
```

```python
dog1.ladrar()
```

```python
dog2 = Dog("Cheese", 2)
```

```python
dog2.ladrar()
```

```python
class Calculadora:

  def __init__(self ,model):
     self.Model = model

  def getModel(self):
    return f"Calculadora marca {self.Model}"

  def sumar(self,a,b):
    return a+b
  
  def multiplicar(self,a,b):
    return a*b
```

```python
calculadora1 = Calculadora("Casio")
```

```python
calculadora1.getModel()
```

```python
calculadora1.sumar(45,12)
```

```python
calculadora1.multiplicar(45,12)
```
