 Haciendo un Simil de la diferencia entre un método en Java y Python


```
 boolean InOrder(int x, int y, int z)
  {
     if (x <= y && y <= z)
	      return true;
     else
	      return false;
  }
```
Puede ser colapsado a:

```
public boolean InOrder(int x, int y, int z){if (x <= y && y <= z) return true; else return false;}
```

Pero en Python la sintaxis obliga a usar la correcta identación dado que no existe las aperturas con llaves o fin de línea de código con ";"

Por ejemplo , esta identación es correcta 

```
def InOrder(x,y,z):
        if (x<=y and y<=z):
            return True
        else:
            return False
```
y esta no:

```
def InOrder(x,y,z): if (x<=y and y<=z): return True else: return False
```

````python
from random import randint as randint

def devuelve_resultado_tres_dados():
  """
    Esta función devuelve como tupla tres valores aleatorios entre 1 y 6
  """ 

  return (randint(1,6),randint(1,6),randint(1,6))
````

````python
resultado = devuelve_resultado_tres_dados()
````

````python
d1 = resultado[0]
d2 = resultado[1]
d3 = resultado[2]
````

````python
print(f"El resultado de los dados es : {d1} {d2} {d3}")
````

````python
d1, d2, d3 = devuelve_resultado_tres_dados()
print(f"El resultado de los dados es : {d1} {d2} {d3}")
````
